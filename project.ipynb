### Cleaning the data

All five sources of data used are in a format that has more than what is needed for this hypothesis. <br><br>
Below two examples will be shown of the uncleaned data. First all the corresponding modules will be uploaded and a function will be written to convert .csv datasets into matrices. 

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as stats
%pylab inline

# converts .csv file to matrix
def convertCSV(file):
    matrix = []
    
    with open(file, 'rt') as csvfile:
        reader=csv.reader(csvfile, delimiter=',')
        matrix = [row for row in reader]
        
    return matrix

# create all matrices from corresponding datasets
marriageData = convertCSV("Marriages.csv")
singleHomeData = convertCSV("One-Person_Household.csv")
gdpData = convertCSV("GDP.csv")
under15Data = convertCSV("Pop15Under.csv")
over65Data = convertCSV("Pop65Older.csv")
    
```

##### Cleaning: GDP per capita dataset
&emsp;The data corresponding to the GDP shows information about all countries versus that of just Japan. It also starts from 1960, however our remaining datasets start from 1975. There are also table elements which are unnecessary including "Country Code", "Indicator Name", and "Indicator Code". <br>
<br>_To clean the GDP dataset:_<br>
   * Only the row with Japan's data will be extracted<br>
   * Only the columns regarding the year will be extracted <br>
   * Only the years from 1975-2013 will be extracted. <br>
   * 'NA' will be assigned to any possible missing rows <br>
   * The GDP values will be converted to two decimal places <br>
   * The rows and columns of the table will then be labeled to match the corresponding data.<br>

```python
# stores Japan's data
japanGDP = []

# find Japan in dataset, store value
for row in gdpData:
    if row[0] == "Japan":
       japanGDP = row

# extract columns 19 to 60 (1975-2013)
japanGDP = japanGDP[19:58]

# convert values to two decimal places
for index in range(len(japanGDP)):
    japanGDP[index] = float(format(float(japanGDP[index]), '.2f'))

# create dataframe, set corresponding row and column values
gdpFrame = pd.DataFrame(japanGDP)
gdpFrame.columns = ['GDP Per Capita']
gdpFrame.index = range(1975, 2014)
gdpFrame.index.name = 'Year'
```

##### Cleaning: Marriage, one-person household, and population datasets
&emsp;The remaining four datasets have identical format so a function will be created to clean the datasets the same way. The only difference will be with the one-person home data because that was recorded every 5 years versus every year. This will result in a table with less rows than the remaining three. These datasets show information for each region of Japan instead of the country as a whole. <br>
<br>_To clean the remaining datasets:_ <br>
* Each region's value will be combined to sum the total for the entire country per year. 
* The extra table elements "region", "indicator", and "Unit" will all be removed
* 'NA' will be assigned to any possible missing rows <br>
* The rows and columns of the table will then be labeled to match the corresponding data.<br>

```python
# cleans datasets from knoema.com
# Parameters:
#    dataset: dataset to clean
#    yearIncrement: how many years each dataset is apart
def cleanKnoemaData(dataset, yearIncrement):
    
    # all years in dataset
    years = {}
    
    # creates list of years according to dataset
    if yearIncrement == 1:
        for year in range(1975,2014):
            years[year] = 0
    else:
        for year in range(1975,2011,5):
            years[year] = 0
       
    # indices of year and value in table
    yearIndex = 3
    valueIndex = 4
    
    # search dataset for each year
    for year in years:
        
        # sum values of all the same years
        for row in range(0,len(dataset)):
            
            # years match
            if dataset[row][yearIndex] == str(year):
                
                #  sum total for corresponding year in dictionary
                years[year] += int(dataset[row][valueIndex])
        
    # create sorted list of years and values
    cleanedData = []
    for year in sorted(years):
        cleanedData.append( [year,years[year]])
    
    # save rows and columns then create dataframe
    cleanColumns = [i[0] for i in cleanedData]
    cleanRows = [i[1] for i in cleanedData]
    frame = pd.DataFrame(cleanRows)
    frame.index = cleanColumns
    frame.index.name = "Year"
        
    return frame
```

### Results: Analyzing the data
&emsp; To begin analyzing the cleaned data different variables will be compared then their correlations and probability values will be calculated. The first focus will be comparing the variables that represent population and lifestyles of residents of Japan, it will then be compared to the GDP to see if these lifestyle choices effect the workforce.

##### First analysis: Population
Below is a graph illustrating the data corresponding to the age demographic of elderly 65 and over compared to youth 15 and under.

```python
# Population and single homes
under15Frame['Total 15 and Under'].plot(legend=True).set(yLabel='Total')
over65Frame['Total 65 and Over'].plot(legend=True)
plt.title('Demographic 65 and older vs. 15 and Under')
```

Eyeballing this graph its clear to see an incline in the population of the elderly and single homes while seeing a decline in the population of the youth. To get a more solid result the correlation coefficient and probability value will be calculated for Japan's population of elderly 65 and over vs. that of the youth 15 and under. 

```python
# linear regression collect data
data65 = over65Frame["Total 65 and Over"]
data15 = under15Frame["Total 15 and Under"]

rho, pvalue = stats.pearsonr(data15, data65)
print ('Correlation Coefficient: ', rho)
print ('Probability Value: ', pvalue)

plt.plot(data65, data15,'.',color='red', alpha=1)
plt.xlabel("Population 65 and Over")
plt.ylabel("Population 15 and Under")
plt.title("15 and Under vs. 65 and Older")
```

As we can see the correlation coefficient is -0.954015474936, almost -1 which means the variables are anti-correlated. This tells us an increase in Japan's population of the elderly is correlated with the decrease in youth aged 15 and under. The P-value is also extremely small  5.95307478622e-21 which is close to 0, this tells us that it is very imporbable that this correlation was by chance. This makes it clear that in Japan they are have much less children than they used to, without a doubt a change in lifestyle is occuring. 

##### Second analysis: Marriages and Single-Homes
This analysis will compare the number of marriages and single households in Japan. If our hypothesis is correct we should see a decrease in the number of marriages and an increase in the number of single households. Below are two graphs to illustrate what these datasets imply.

```python
# Marraiges
marriagesPlot = marriageFrame.plot(lw=2,colormap='jet',marker='.',markersize=10)
marriagesPlot.set(xlabel='Year', ylabel="Total Marriages")
plt.title('Total Marriages vs. Year')

# Single homes
singleHomePlot = singleHomeFrame.plot(lw=2,colormap='jet',marker='.',markersize=10,title="Ratio of One-Person Households")
singleHomePlot.set(xlabel='Year', ylabel="One-Person Homes")
```

Next the correlation coefficient and probability value will be calculated. To do so the marriage data must unexpectedly be cleaned further to be compared every 5 years instead of every year to match the single household data.

```python
# marriages data reduced number of columns to match Singe Households data
marriageFrameReduced = cleanKnoemaData(marriageData, 5)
marriageFrameReduced.columns = ["Marriages"]

# linear regression collect data
dataMarriages = marriageFrameReduced["Marriages"]
dataHouseholds = singleHomeFrame["One-Person Homes"]

rho, pvalue = stats.pearsonr(dataMarriages, dataHouseholds)
print ('Correlation Coefficient: ', rho)
print ('Probability Value: ', pvalue)

plt.plot(dataMarriages, dataHouseholds,'.',color='blue', alpha=1)
plt.xlabel("Total Marriages")
plt.ylabel("Single Households")
plt.title("Marriages vs. Single Households")
```

The correlation coefficient is -0.641751489657, this is close to -1 showing anti-correlation. Looking at the scatter plot we can clearly see as the single households increase the number of marriages decrease. The P-value is 0.0862743008834 which is very small and close to 0 showing that this correlation is unlikely to have occured by chance.

The results show a negative correlation (-0.511223802283) with a low probability value (0.000881232147204) which is what our hypothesis suggests.
